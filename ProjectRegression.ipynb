# **Load ML Project**

#**Load Data**

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/dataprofessor/data/master/delaney_solubility_with_descriptors.csv')
```

# **Data Preparation**

## **Data separation as X and Y**

```python
y = df['logS']
y
```

```python
x = df.drop('logS', axis =1)
x
```

## **data split**

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=100)
```

```python
x_train
```

```python
x_test
```

# Model building

## Linear regressoion

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
y_lr_train_pred = lr.predict(x_train)
y_lr_test_pred = lr.predict(x_test)
```

```python
y_lr_test_pred
```

```python
from sklearn.metrics import mean_squared_error, r2_score
lr_train_mse = mean_squared_error(y_train, y_lr_train_pred)
lr_train_r2 = r2_score(y_train, y_lr_train_pred)

lr_test_mse = mean_squared_error(y_test, y_lr_test_pred)
lr_test_r2 = r2_score(y_test, y_lr_test_pred)
```

## **Random Forest**

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
rf.fit(x_train, y_train)
y_rf_train_pred = rf.predict(x_train)
y_rf_test_pred = rf.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
rf_train_mse = mean_squared_error(y_train, y_rf_train_pred)
rf_train_r2 = r2_score(y_train, y_rf_train_pred)

rf_test_mse = mean_squared_error(y_test, y_rf_test_pred)
rf_test_r2 = r2_score(y_test, y_rf_test_pred)
```

```python
lr_result = pd.DataFrame([
    ["Linear regression", lr_train_mse, lr_train_r2, lr_test_mse, lr_test_r2],
    ["Random forest", rf_train_mse, rf_train_r2, rf_test_mse, rf_test_r2]
], columns = ["Method", "Train MSE", "Tarin R2", "Test MSE", "Test R2"])
lr_result
```

```python
from matplotlib import pyplot as plt
import numpy as np

z = np.polyfit(y_train, y_lr_train_pred,1)
p = np.poly1d(z)

plt.plot(y_train, p(y_train), '#ff0000')
plt.scatter(y_train, y_lr_train_pred, c="#1101", alpha=0.3)
plt.xlabel("Actual logS")
plt.ylabel("Predicted logS")
```
